```python
# Drop Duplicates Data
print("Total Data Before Drop Duplicates: ", df_train.shape[0])
df_train = df_train.drop_duplicates()
print("Total Data After Drop Duplicates: ", df_train.shape[0])
```

```python
# Load Data Test
df_test = pd.read_csv("./UCI HAR Dataset/test/X_test.txt", header=None, 
                 sep="\s+")
df_test.columns = list(df_features.squeeze().values)
df_test_labels = pd.read_csv("./UCI HAR Dataset/test/y_test.txt", header=None)
df_test["Activity"] = df_test_labels[0]
df_test["Activity"] = df_test["Activity"].apply(lambda x: x - 1)
df_test.head()
```

```python
# Split Data Train and Data Test
x_train = df_train.drop(columns=["Activity"])
y_train = df_train[["Activity"]]
x_val = df_test.drop(columns=["Activity"])
y_val = df_test[["Activity"]]
print("Total Features Train: ", x_train.shape[0])
print("Total Labels Train: ", y_train.shape[0])
print("Total Features Test: ", x_val.shape[0])
print("Total Labels Test: ", y_val.shape[0])
```

```python
# Make Model
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, activation="relu", 
                          input_shape=[x_train.shape[1]]),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(256, activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(512, activation="relu"),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dense(y_train["Activity"].unique().shape[0],
                          activation='softmax')
])

model.compile(
    optimizer="Adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
# Make Callbacks For Deep Learning
callback0 = tf.keras.callbacks.EarlyStopping(
    min_delta=0.001,
    patience=8,
    restore_best_weights=True
)
print("Callback Ready")
```

```python
# Train Model
history = model.fit(
    x_train, y_train.values,
    validation_data=(x_val, y_val.values),
    epochs=100,
    callbacks=[callback0]
)
```

```python
# Plot Accuracy Model
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Accuracy Model')
plt.legend(['accuracy', 'val_accuracy'])
plt.show()
```

```python
# Plot Loss Function Model
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss Function Model')
plt.legend(['loss', 'val_loss'])
plt.show()
```
